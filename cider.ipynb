## Properties 


### NCPR - Get the net charge per residue of the sequence 
Polyelectrolytic IDPs with NCPR below a threshold value of 0.25 adopt compact globular ensembles, whereas sequences that lie above this threshold adopt well-solvated expanded coils and even stiff rod-like conformations. 

### Disorder promoting 

Get the fraction of residues predicted to be ‘disorder promoting’ [1]. Note this is NOT a disorder prediction!

### Omega 

Get the sequence’s Omega value. Omega defines the patterning between charged/proline residues and all other residues  

```python
objects = []
kappas = []
neg = []
pos = []
charge = []
NCPRs = []
disorder_promoting = []
omegas = [] 
 
for seq in seq_list: 
    SeqOb = SequenceParameters(seq) 
    objects.append(SequenceParameters(seq))
    kappas.append(SeqOb.get_kappa()) 
    neg.append(SeqOb.get_fraction_negative()) 
    pos.append(SeqOb.get_fraction_positive())
    charge.append(SeqOb.get_mean_net_charge(pH=7))
    NCPRs.append(SeqOb.get_NCPR(pH=7))
    disorder_promoting.append(SeqOb.get_fraction_disorder_promoting())
    omegas.append(SeqOb.get_Omega())
```

```python
df = pd.DataFrame(list(zip(name_list, kappas, charge, neg, pos, NCPRs, disorder_promoting, omegas)),
               columns =["Protein", "Kappa", "Charge","Fraction Neg", "Fraction Pos", "NCPR", "Disorder promoting", "Omega"])
df
```

```python
# showing with sequence labels
plots.show_multiple_phasePlot2(objects,label_list=name_list, fontSize=3)

# saving without
plots.save_multiple_uverskyPlot2(objects, 'Uversky MTMR family', 
                                 name_list, fontSize=3)
plots.save_multiple_phasePlot2(objects, 'Phase MTMR family', name_list, 
                               fontSize=3)
```
